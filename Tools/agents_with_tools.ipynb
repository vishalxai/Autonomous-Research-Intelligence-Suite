```python
import os
from dotenv import load_dotenv
from autogen_agentchat.agents import AssistantAgent
from autogen_ext.models.openai import OpenAIChatCompletionClient
from autogen_agentchat.messages import TextMessage
from autogen_agentchat.ui import Console
from autogen_core import CancellationToken

# 1. Load the secret key securely
load_dotenv()
or_key = os.getenv("OPENROUTER_API_KEY")

print("✅ Block 1 Complete: Imports loaded and API Key retrieved.")
```

```python
# 2. Configure the Cloud Brain (Auto-Routing)
cloud_model_client = OpenAIChatCompletionClient(
    model="openrouter/free", # Automatically routes to the best available free model!
    api_key=or_key,
    base_url="https://openrouter.ai/api/v1",
    model_info={
        "vision": False,
        "function_calling": True,
        "json_output": True,
        "family": "unknown", # Set to unknown because it could route to Llama, Mistral, etc.
    }
)

print("✅ Block 2 Complete: Auto-Routing Cloud Client configured.")
```

```python
# 3. Create a Cloud-Powered Agent
cloud_agent = AssistantAgent(
    name="cloud_researcher",
    model_client=cloud_model_client,
    system_message="You are a high-level research architect. Provide deep, expert-level insights."
)

print("✅ Block 3 Complete: Cloud Agent Persona created.")
```

```python
# 4. Run the Cloud Task with Resiliency
async def main():
    print("--- Connecting to Cloud Brain ---")
    try:
        await Console(
            cloud_agent.on_messages_stream(
                messages=[TextMessage(content="Compare the architectural benefits of Multi-Agent systems vs Single LLM chains in 2 concise bullet points.", source="User")],
                cancellation_token=CancellationToken()
            )
        )
    except Exception as e:
        # If OpenRouter is full, catch the error so the app doesn't crash
        print(f"\n⚠️ Cloud API Congested (Error 429): {e}")
        print("🔄 Fallback Triggered: In a production app, we would now automatically route this task back to your local Llama 3.2 on the M1 Mac!")

# THE JUPYTER FIX: Use await instead of asyncio.run
await main()
```
