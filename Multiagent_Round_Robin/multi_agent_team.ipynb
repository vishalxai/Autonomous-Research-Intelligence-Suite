```python
import os
from dotenv import load_dotenv
from autogen_agentchat.agents import AssistantAgent
from autogen_ext.models.openai import OpenAIChatCompletionClient
from autogen_agentchat.teams import RoundRobinGroupChat
from autogen_agentchat.ui import Console

load_dotenv()
or_key = os.getenv("OPENROUTER_API_KEY")
```

```python
# 1. The Cloud Brain (Auto-Routing Load Balancer)
# Since this is a text task, the router will dynamically bypass all 429 traffic jams!
cloud_client = OpenAIChatCompletionClient(
    model="openrouter/free", 
    api_key=or_key,
    base_url="https://openrouter.ai/api/v1",
    model_info={"vision": False, "function_calling": True, "json_output": False, "family": "unknown"}
)
```

```python
# 2. The Local Brain (Fast Processing/Formatting)
local_client = OpenAIChatCompletionClient(
    model="llama3.2",
    api_key="placeholder",
    base_url="http://localhost:11434/v1",
    model_info={"vision": False, "function_calling": True, "json_output": False, "family": "llama3"}
)

print("✅ Block 1 & 2: Cloud Brain swapped to the OpenRouter Free Load-Balancer!")
```

```python
# 3a. The Cloud Researcher (Powered by Llama 3.3 70B)
cloud_researcher = AssistantAgent(
    name="Cloud_Researcher",
    model_client=cloud_client,
    system_message="You are a brilliant agricultural researcher. When asked a question, provide a detailed, 2-paragraph analysis."
)

# 3b. The Local Formatter (Powered by Local Llama 3.2)
local_formatter = AssistantAgent(
    name="Local_Formatter",
    model_client=local_client,
    system_message="You are a strict data formatter. Take the researcher's output and condense it into exactly 3 concise bullet points. Do not add any conversational filler like 'Here are the points'."
)

print("✅ Block 3: Agents Created with Specialized Personas")
```

```python
# 4. Create the Hybrid Team
hybrid_team = RoundRobinGroupChat(
    participants=[cloud_researcher, local_formatter],
    max_turns=2 # Turn 1: Cloud researches. Turn 2: Local formats.
)

# 5. Execute the Team Workflow (With Resilient Error Catching)
async def run_hybrid_team():
    print("--- 🚀 Starting Hybrid Multi-Agent Execution ---")
    task = "Analyze the primary benefits of precision agriculture using drone technology."
    
    try:
        # Run the stream so we can watch them talk to each other in real-time
        await Console(
            hybrid_team.run_stream(task=task)
        )
    except Exception as e:
        print(f"\n⚠️ Cloud API Congested: {e}")
        print("🔄 Wait 60 seconds and run this cell again. If it continues to fail, swap the Cloud Brain to 'google/gemma-3-27b-it:free' in Block 1.")

# THE JUPYTER FIX
await run_hybrid_team()
```
