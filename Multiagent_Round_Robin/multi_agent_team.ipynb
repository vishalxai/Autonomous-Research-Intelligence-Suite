```python
import os
from dotenv import load_dotenv
from autogen_agentchat.agents import AssistantAgent
from autogen_ext.models.openai import OpenAIChatCompletionClient
from autogen_agentchat.teams import RoundRobinGroupChat
from autogen_agentchat.ui import Console

load_dotenv()
or_key = os.getenv("OPENROUTER_API_KEY")
```

```python
# 1. The Cloud Brain (Auto-Routing Load Balancer)
# Since this is a text task, the router will dynamically bypass all 429 traffic jams!
cloud_client = OpenAIChatCompletionClient(
    model="meta-llama/llama-3.3-70b-instruct",
    api_key=or_key,
    base_url="https://openrouter.ai/api/v1",
    model_info={"vision": False, "function_calling": True, "json_output": False, "family": "unknown"}
)
```

```python
# 2. The Local Brain (Fast Processing/Formatting)
local_client = OpenAIChatCompletionClient(
    model="llama3.2",
    api_key="placeholder",
    base_url="http://localhost:11434/v1",
    model_info={"vision": False, "function_calling": True, "json_output": False, "family": "llama3"}
)

print("✅ Block 1 & 2: Cloud Brain swapped to the OpenRouter Free Load-Balancer!")
```

```python
import subprocess

def search_chub_docs(query: str) -> str:
    """
    Search Context Hub (chub) for the latest 2026 API documentation and SDK standards.
    Use this tool whenever you need to verify model names (like gpt-5.2) or correct SDK syntax.
    """
    try:
        # Executes the terminal command 'chub search' and captures the output
        result = subprocess.run(['chub', 'search', query], capture_output=True, text=True)
        if result.returncode == 0:
            return result.stdout
        else:
            return f"Error searching chub: {result.stderr}"
    except Exception as e:
        return f"Failed to execute chub search: {str(e)}"

print("✅ Step 2: Documentation Search Tool Defined")
```

```python
# 3a. The Cloud Researcher (Now Forced to Reflect)
cloud_researcher = AssistantAgent(
    name="Cloud_Researcher",
    model_client=cloud_client,
    tools=[search_chub_docs],
    reflect_on_tool_use=True, # 🛑 THE FIX: Forces the agent to read the tool output and speak!
    system_message=(
        "You are a Senior AI Engineer. Your goal is to provide deep research and code. "
        "CRITICAL: If you are generating code or referencing model specs, you MUST use the "
        "'search_chub_docs' tool first to ensure you are using the latest 2026 standards. "
        "RULE: After executing a tool, you MUST output a final, comprehensive text response "
        "synthesizing the tool's output and answering the user's original prompt."
    )
)

print("✅ Step 3 Updated: Cloud Researcher forced to reflect on tool use.")

# 3b. The Local Formatter (Stays the same)
local_formatter = AssistantAgent(
    name="Local_Formatter",
    model_client=local_client,
    system_message="You are a strict data formatter. Condense the research into 3 concise bullet points."
)

print("✅ Step 3: Cloud Researcher Equipped with Documentation Tool")
```

```python
from autogen_agentchat.agents import UserProxyAgent
from autogen_agentchat.teams import RoundRobinGroupChat

# 3c. The Human Boss (You)
user_proxy = UserProxyAgent(
    name="Human_Admin",
    description="A human supervisor who reviews research and provides final approval."
)

# 4. Create the Supervised Team (Replacing 'hybrid_team')
supervised_team = RoundRobinGroupChat(
    participants=[cloud_researcher, local_formatter, user_proxy], 
    max_turns=6 # Turn 1: Cloud researches, Turn 2: Local formats, Turn 3: YOU review
)

print("✅ Step 4 Updated: Supervised Team and Human Proxy defined!")

# 5. The Integrated Execution (Agriculture + Technical Precision)
async def run_integrated_suite():
    print("--- 🌾 Starting Integrated Research & Verification ---")
    
    # This task combines your MAIN purpose with your NEW technical power
    task = (
        "Part 1: Analyze the primary benefits of precision agriculture using drone technology. "
        "Part 2: Use your 'search_chub_docs' tool to find the 2026 standard for 'openai/vision' "
        "and explain how it could be used to analyze the drone imagery described in Part 1."
    )
    
    try:
        # This will run the Researcher -> Formatter -> You (Human_Admin)
        await Console(
            supervised_team.run_stream(task=task)
        )
    except Exception as e:
        print(f"⚠️ System Error: {e}")

# TRIGGER THE FULL SUITE
await run_integrated_suite()
```
